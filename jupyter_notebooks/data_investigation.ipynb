### Column Summary

```python
# Summary of all columns: data type, nulls, unique values, example values
summary = pd.DataFrame({
    'dtype': df.dtypes,
    'nulls': df.isnull().sum(),
    'unique_values': df.nunique(),
    'example_values': df.apply(lambda x: x.unique()[:5])
})

summary
```

Check unique brands

```python
df['CarBrand'].value_counts()
```

Check for Duplicates in Brands

```python
df["CarBrand"] = df["CarBrand"].replace({
    "vw" : "volkswagen",
    "vokswagen" : "volkswagen",
    "porcshce" : "porsche",
    "maxda" : "mazda",
    "toyouta" : "toyota"
})
```

```python
df["CarBrand"].value_counts()
```

Review numeric values

```python
df.describe().T
```

```python
df[['CarBrand', 'price']].sort_values(by='price', ascending=False).head()
```

```python
# Convert symboling to categorical since it's not a continuous variable
df['symboling'] = df['symboling'].astype('category')
df.head()
```

Clean Text Data in Categorical Columns

```python
#Identify all text columns (type = object)
cat_cols = df.select_dtypes(include='object').columns

#Loop through each and apply text cleaning
for col in cat_cols:
    df[col] = df[col].str.lower().str.strip()

# Check the cleaned values
for col in cat_cols:
    print(f"Unique values in {col}: {df[col].unique()}\n")
```

Review Numeric Columns for Outliers

```python
# Check for very low prices or invalid values
df[df["price"] < 1000]

# Cars with very low horsepower
df[df["horsepower"] < 40]

# Cars with very small engine size
df[df["enginesize"] < 50]
```

```python
df[["CarBrand", "price"]].sort_values(by="price", ascending=False).head()
```

I added derived metrics such as price per horsepower and power-to-weight ratio to better understand performance-to-cost characteristics across car brands.

```python
# Create new columns based on existing ones
df['price_per_hp'] = df['price'] / df['horsepower']
df['power_to_weight'] = df['horsepower'] / df['curbweight']
df.head(5)
```

### Convert All Column Names to Lowercase

```python
df.columns = df.columns.str.lower()
```

Confirm categorical and numerical columns look right

```python
cat_cols = df.select_dtypes(include="object").columns
num_cols = df.select_dtypes(include=["int64", "float64"]).columns

print("Categorical:", cat_cols.tolist())
print("Numerical:", num_cols.tolist())
```

Convert Categorical Columns to category Type

```python
categorical_cols = ["fueltype", "aspiration", "doornumber", "carbody", "drivewheel", "enginelocation", "enginetype", "cylindernumber", "fuelsystem", "carbrand"]

for col in categorical_cols:
    df[col] = df[col].astype("category")

df.info()
```
